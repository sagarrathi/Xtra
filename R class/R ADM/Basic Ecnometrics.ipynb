```r
curve(dchisq(x, df=3), 
    xlim=c(0,10),
    ylim=c(0,1),
    col="blue",
    ylab="",
    main="PDF & CDF")

curve(pchisq(x, df=3),
    xlim=c(0,10),
    add=T,
    col="red")

legend("topleft", 
      c("PDF", "CDF"),
      col=c("blue", "red"),
      lty=c(1,1) 
      )
```

```r
curve(dchisq(x, df=1),
      xlim =c(0, 15),
      xlab="x",
      ylab="Density",
      main="CHI SQUARE Distribution",
      col=
    )

for (M in 2:7) {
    curve(dchisq(x, df=M),
    xlim=c(0, 15),
    add =T,
    col =M)
}


legend ("topright", 
       as.character(1:7), 
       col=1:7,
       lty=1,
       title="D.F.")
```

```r
curve(dnorm(x),
     xlim=c(-4,4),
     xlab="x",
     lty=2,
     ylab="Density",
     main="Density vs T Distrubution")

curve(dt(x, df=2),
     xlim=c(-4,4),
     col=2,
     add=T)

curve(dt(x, df=4),
     xlim=c(-4,4),
     col=3,
     add=T)

curve(dt(x, df=25),
     xlim=c(-4,4),
     col=4,
     add=T)

legend("topright",
      c("N", "M=2","M=4","M=25"),
      col=1:4,
      lty=c(2,1,1,1) 
      )
```
