```python
import pandas as pd
import numpy as np
```

#### Retriving toronto data

```python
tt_df=pd.read_csv("toronto_data.csv")
```

# Gettting Lat & Long of data


#### Setting up geopy


```python
from geopy.geocoders import Nominatim
geolocator=Nominatim(user_agent="coursera_capstone_05", timeout=15)


# import geocoder

tt_df["Latitude"]=""
tt_df["Longitude"]=""
```

#### Updating DF

```python
for i  in tt_df.index: 

    
    neighbourhood=tt_df.iloc[i]["Neighbourhood"]
    borough=tt_df.iloc[i]["Borough"]
    
    neighbourhood=neighbourhood.split(",")[0]
    
    postcode=tt_df.iloc[i]["PostalCode"]
    
    address=neighbourhood+", Toronto, Ontario"
    
    
    print("Address:",address)
    
    
    location=geolocator.geocode(address)
    """
    Geocoder failed as mentioned in coursera
    
    location=None
    while (location==None):
        g=geocoder.google(address)
        location=g.latlang
    """
    
    try:
        lat=location.latitude
        long=location.longitude
    except:
        lat=""
        long=""
        
    tt_df.at[i,"Latitude"]=lat
    tt_df.at[i,"Longitude"]=long


    print("Latitude:",lat)
    print("Longitude:",long)
```

```python
tt_df.head()
```

#### Saving DF

```python
tt_df.to_csv("./toronto.csv", index=False)
```

#### Update those locations with csv file

```python
cord_df=pd.read_csv("Geospatial_Coordinates.csv")
cord_df.head()
```

```python
for i in tt_df.index:
    if pd.isnull(tt_df.iloc[i]["Latitude"]):
        postcode=tt_df.iloc[i]["PostalCode"]
        print("Corrected Location For:",postcode)
        row=cord_df[cord_df["Postal Code"]==postcode]
        lat=row["Latitude"].values
        long=row["Longitude"].values
        
        tt_df.at[i,"Latitude"]=lat
        tt_df.at[i,"Longitude"]=long
tt_df.head()
```
